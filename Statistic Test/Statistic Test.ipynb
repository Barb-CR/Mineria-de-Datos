```python
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import f_oneway, kruskal


df = pd.read_csv("C:/Users/sesa777393/Desktop/tareas/PIA_MIN_DAT/Data Cleaning/dataset_clean.csv")
```

```python
if 'medal' in df.columns and 'usability' in df.columns:
    if df['medal'].nunique() >= 2:
        medal_groups = df.groupby("medal")["usability"].apply(list)

        if len(medal_groups) >= 3:
            try:
                anova_result = f_oneway(*medal_groups)
                print("Resultado ANOVA (Usability por Medal):")
                print("Estadístico F:", anova_result.statistic)
                print("p-valor:", anova_result.pvalue)
            except ValueError as e:
                print(f"Error running ANOVA (Usability por Medal): {e}. Likely not enough data in each group.")
        else:
            print("ANOVA requires at least 3 groups. Skipping ANOVA for Usability by Medal.")

        try:
            kruskal_result = kruskal(*medal_groups)
            print("\nResultado Kruskal-Wallis (Usability por Medal):")
            print("Estadístico H:", kruskal_result.statistic)
            print("p-valor:", kruskal_result.pvalue)
        except ValueError as e:
            print(f"Error running Kruskal-Wallis (Usability por Medal): {e}. Likely not enough data in each group.")


        print("\nConclusión (Usability por Medal):")
        alpha = 0.05
        if len(medal_groups) >= 3 and anova_result.pvalue < alpha:
            print(f"El resultado del ANOVA sugiere que hay una diferencia estadísticamente significativa en la usabilidad promedio entre al menos dos de los grupos de medallas (p < {alpha}).")
        elif kruskal_result.pvalue < alpha:
            print(f"El resultado del Kruskal-Wallis sugiere que hay una diferencia estadísticamente significativa en la distribución de la usabilidad entre al menos dos de los grupos de medallas (p < {alpha}).")
        else:
            print(f"No se encontró una diferencia estadísticamente significativa en la usabilidad entre los grupos de medallas (p >= {alpha}).")

    else:
        print("Not enough unique medal categories to perform group comparison tests.")
else:
    print("The 'medal' or 'usability' column was not found in the DataFrame.")

if 'medal' in df.columns and 'views' in df.columns:
    if df['medal'].nunique() >= 2:
        medal_groups_views = df.groupby("medal")["views"].apply(list)

        if len(medal_groups_views) >= 3:
            try:
                anova_result_views = f_oneway(*medal_groups_views)
                print("\nResultado ANOVA (Views por Medal):")
                print("Estadístico F:", anova_result_views.statistic)
                print("p-valor:", anova_result_views.pvalue)
            except ValueError as e:
                print(f"Error running ANOVA (Views por Medal): {e}. Likely not enough data in each group.")
        else:
            print("ANOVA requires at least 3 groups. Skipping ANOVA for Views by Medal.")

        try:
            kruskal_result_views = kruskal(*medal_groups_views)
            print("\nResultado Kruskal-Wallis (Views por Medal):")
            print("Estadístico H:", kruskal_result_views.statistic)
            print("p-valor:", kruskal_result_views.pvalue)
        except ValueError as e:
            print(f"Error running Kruskal-Wallis (Views por Medal): {e}. Likely not enough data in each group.")

        print("\nConclusión (Views por Medal):")
        alpha = 0.05
        if len(medal_groups_views) >= 3 and anova_result_views.pvalue < alpha:
            print(f"El resultado del ANOVA sugiere que hay una diferencia estadísticamente significativa en el promedio de vistas entre al menos dos de los grupos de medallas (p < {alpha}).")
        elif kruskal_result_views.pvalue < alpha:
            print(f"El resultado del Kruskal-Wallis sugiere que hay una diferencia estadísticamente significativa en la distribución de las vistas entre al menos dos de los grupos de medallas (p < {alpha}).")
        else:
            print(f"No se encontró una diferencia estadísticamente significativa en las vistas entre los grupos de medallas (p >= {alpha}).")
    else:
        print("Not enough unique medal categories to perform group comparison tests for Views.")
else:
    print("The 'medal' or 'views' column was not found in the DataFrame.")
```
