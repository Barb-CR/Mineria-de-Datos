```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/sesa777393/Desktop/tareas/PIA_MIN_DAT/Data Cleaning/dataset_clean.csv")
```

```python
numeric_cols = ['usability', 'views', 'downloads']
plt.figure(figsize=(15, 5))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, 3, i)
    df[col].hist(bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(df['views'], df['downloads'], color='coral', alpha=0.5)
plt.title('Views vs. Downloads')
plt.xlabel('Views')
plt.ylabel('Downloads')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
medal_counts = df['medal'].value_counts()
plt.pie(medal_counts, labels=medal_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Medal Distribution')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
types_of_files_counts = df['types_of_files'].value_counts().head(10)  # Top 10 types
types_of_files_counts.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Types of Files')
plt.xlabel('Type of File')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
