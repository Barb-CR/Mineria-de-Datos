```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/sesa777393/Desktop/tareas/PIA_MIN_DAT/Data Cleaning/dataset_clean.csv")
```

```python
df
```

```python

print(df.describe())
print(df.columns)
```

```python
print("\n--- Dataset Statistics Grouped by Medal ---")
print(df.groupby('medal')[['views', 'downloads', 'usability', 'num_of_files']].agg(['mean', 'median', 'count']))

```

```python
print("\n--- Dataset Statistics Grouped by File Type ---")
print(df.groupby('types_of_files')[['views', 'downloads', 'usability']].agg(['mean', 'median', 'count']))

```

```python

print("\n--- Statistics for Specific Numeric Columns ---")
print("Mean usability:", df['usability'].mean())
print("Median views:", df['views'].median())
print("Total downloads:", df['downloads'].sum())
```

```python
print("\n---Correlation between numeric columns---")
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
print(corr_matrix)
```

```python
print("\n--- Medal Impact on Engagement ---")
print(df.groupby('medal')[['views', 'downloads', 'usability']].mean())

plt.figure()
df.plot.scatter(x='views', y='downloads')
plt.title('Views vs. Downloads')
plt.show()
```
